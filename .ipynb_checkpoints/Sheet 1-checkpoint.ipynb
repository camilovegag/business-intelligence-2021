```python
# 2.

M = np.array([y, y2])

# first way
M2 = np.transpose(M)

# second way
M2 = np.moveaxis(M, 0, 1)
```

```python
# 1.

import pandas as pd
columns = ['Name', 'Age', 'Sex', 'Profession']

df1 = pd.DataFrame({
    "Name":["Alexander","Jane",],
    "Age":[19, 36],
    "Sex": ["male", "female"],
    "Profession": ["student", "physician"],
}, columns = columns)
df2 = pd.DataFrame({
    "Name":["Eric", "Laura",],
    "Age":[22, 48],
    "Sex": ["male", "female"],
    "Profession": ["lawyer", "teacher"],
}, columns = columns)
df3 = pd.DataFrame({
    "Name":["Peter", "Julia",],
    "Age":[31, 24],
    "Sex": ["male", "female"],
    "Profession": ["engineer", "consultant"],
}, columns = columns)
```

```python
# 2. 

union = [df1, df2, df3]
df = pd.concat(union, ignore_index=True)
df
```

```python
# 3.

new_df = pd.DataFrame({
    "Name": ["Alexander", "Jane", "Eric", "Laura", "Peter", "Julia"],
    "Size": [181, 162, 178, 183, 173, 169],
    "Salary":[600, 4200, 4000, 3400, 5200, 3600]
})
new_df
joined_df = df.join(new_df.set_index('Name'), on="Name")
joined_df
```

```python
# 4.

joined_df['Size'] -= 1
joined_df
```

```python
# 5.

# first way
joined_df[joined_df.Age > 30]

# second way
joined_df.query('Age > 30')
```

```python
# 6.

joined_df.sort_values('Salary')
```

```python
# 7.

males = joined_df.query("Sex == 'male'")
females = joined_df.query("Sex == 'female'")

print(males)
print(females)
```
